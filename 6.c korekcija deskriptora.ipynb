```python
import pandas as pd
```

```python
from sklearn.experimental import enable_iterative_imputer  # Ovo je potrebno za aktivaciju
from sklearn.impute import IterativeImputer
```

```python
desc = pd.read_csv("descriptors2.csv")
desc
```

```python
# Kolone koje imaju barem jedan NaN
nan_counts = desc.isna().sum()
nan_counts = nan_counts[nan_counts > 0]
nan_counts
```

```python
#Iterative Imputer, fit-transformacija
desc_numeric = desc.select_dtypes(include=['float64', 'int64'])

imputer = IterativeImputer(random_state=0)

desc_imputed = imputer.fit_transform(desc_numeric)

desc_imputed_df = pd.DataFrame(desc_imputed, columns=desc_numeric.columns)

desc_clean = desc.copy()
desc_clean[desc_numeric.columns] = desc_imputed_df
```
