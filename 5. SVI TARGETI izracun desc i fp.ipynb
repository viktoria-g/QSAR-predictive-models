```python
from rdkit import Chem
from mordred import Calculator, descriptors
import pandas as pd
```

```python
stand_rename_col=pd.read_csv("stand_rename_col.csv")
```

```python
stand_rename_col
```

```python
# Konverzija u RDKit molekule
mol_list = [Chem.MolFromSmiles(s) for s in stand_rename_col['smiles_isomeric']]
```

```python
calc = Calculator(descriptors, ignore_3D=True)
desc_df = calc.pandas(mol_list)
```

```python
# Spajanje po indeksima
rezultat = pd.concat([stand_rename_col.reset_index(drop=True), desc_df], axis=1)
```

```python
rezultat
```

```python
rezultat_0_1 = rezultat.replace({
    "Active": 1,
    "Inactive": 0
})
rezultat_0_1
```

```python
rezultat_0_1.to_csv("desc+ostalo.csv", index=False)
```

```python
#IZRAČUN FINGERPRINTEVA
from rdkit.Chem import AllChem
import numpy as np
```

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
def calculate_morgan_fp(smiles, radius=2, nBits=5120):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return np.nan
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=nBits)
    arr = np.zeros((nBits,), dtype=int)
    AllChem.DataStructs.ConvertToNumpyArray(fp, arr)
    return arr
```

```python
fps = rezultat_0_1['smiles_isomeric'].apply(calculate_morgan_fp)
fps
```

```python
fp_df = pd.DataFrame(fps.tolist())
fp_df.columns = [f'fp_{i}' for i in range(5120)]
```

```python
sve_desc_fp = pd.concat([rezultat_0_1.reset_index(drop=True), fp_df], axis=1)
sve_desc_fp
```

```python
sve_desc_fp.to_csv("sve_desc_fp.csv",index=False)
```
