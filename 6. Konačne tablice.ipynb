```python
#RIJEŠAVANJE DUPLIKATA (ako postoje)
def checkDouble(df, column):

    smi_doub_list = df[column][df[column].duplicated()==True].tolist()

    print('\nDouble smiles report:\nAmount of doubles: %s\n' %len(smi_doub_list))

    return smi_doub_list


smi_doub_index_test = checkDouble(all_data[all_data['set']=='test'], 'smiles_isomeric')
```

```python
def AggSameness(x):
    """
    Provjerava jesu li prve dvije vrijednosti u seriji jednake.
    """
    if len(x) >= 2 and x.values[0] == x.values[1]:
        return True
    else:
        return False

def AggShowBoth(x):
    """
    Vraća tuple s prvim dvjema vrijednostima.
    """
    if len(x) >= 2:
        return (x.values[0], x.values[1])
    elif len(x) == 1:
        return (x.values[0], None)
    else:
        return (None, None)

def AggRemDupl(x):
    """
    Ako su vrijednosti jednake, vraća jednu.
    Ako su obje NaN, vraća NaN.
    Inače, vraća listu obje vrijednosti.
    """
    if len(x) < 2:
        return x.values[0] if len(x) == 1 else np.nan

    v1, v2 = x.values[0], x.values[1]

    if pd.isna(v1) and pd.isna(v2):
        return np.nan
    elif v1 == v2:
        return v1
    else:
        return [v1, v2]

def AggGetOne(x):
    """
    Vraća najveću vrijednost iz serije, ignorira NaN.
    Ako je sve NaN, vraća NaN.
    """
    return x.dropna().max()
```

```python
from rdkit import Chem
from rdkit.Chem import Descriptors
from rdkit.Chem import PandasTools
```

```python
all_corrected_indices["rdkit_mol"] = all_corrected_indices["smiles_isomeric"].apply(Chem.MolFromSmiles)
```

```python
# Lista svih RDKit deskriptora (ime i funkcija)
desc_list = Descriptors.descList
```

```python
# Izračun svih deskriptora i spremanje u novi DataFrame
rdkit_desc = pd.DataFrame([
    [func(mol) if mol is not None else None for name, func in desc_list]
    for mol in all_corrected_indices["rdkit_mol"]
], columns=[name for name, func in desc_list])

rdkit_desc
```

```python
# Kombiniranje s originalnim podacima 
desc_all = pd.concat([all_corrected_indices.reset_index(drop=False), rdkit_desc], axis=1)
desc_all
```

```python
# Lista deskriptora koje želimo ukloniti
desc_to_remove = [
    'SMR_VSA8', 'SlogP_VSA9', 'fr_prisulfonamd',
    'MaxAbsPartialCharge', 'MinPartialCharge',
    'MaxPartialCharge', 'MinAbsPartialCharge'
]
```

```python
# Filtriramo listu samo na one koje postoje u desc_all
existing_cols_to_drop = [col for col in desc_to_remove if col in desc_all.columns]
```

```python
# Uklonimo ih
desc_all = desc_all.drop(columns=existing_cols_to_drop)
desc_all
```

```python
desc_all = desc_all.rename(columns={"smiles_isomeric": "smiles"})
desc_all
```

```python
desc_all = desc_all.drop(columns=["set", "CASRN", "smiles"])
desc_all
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
import warnings
```

```python
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
# Funkcija za dobivanje otiska
def get_morgan_fp(mol, radius=2, nBits=5120):
    if mol is None:
        return [np.nan] * nBits
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=radius, nBits=nBits)
    return list(fp)
```

```python
# Generiraj otiske
fp_array = all_corrected_indices["rdkit_mol"].apply(get_morgan_fp)

# Pretvori u DataFrame
fp_df = pd.DataFrame(fp_array.tolist(), index=all_corrected_indices.index)
```

```python
#Preimenovanje kolona
fp_df.columns = [f'fp{i}' for i in range(5120)]
```
